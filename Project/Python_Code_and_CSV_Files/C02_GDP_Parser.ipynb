## Python CSV parser program for C02 GDP 
#### This program helps to clean the csv files being used for the DV project and only keeps the necessary amout to data and store it in a new CSV file

```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('C02_GDP.csv',skiprows=2,names=['Region','CountryName', 'Year', 'Series','Value']  ,usecols=[0,1,2,3,4])
```

```python
df.head(10)
```

```python
df.replace("GDP per capita (US dollars)",np.nan,inplace=True)
df.replace("GDP in constant 2010 prices (millions of US dollars)",np.nan,inplace=True)
df.replace("GDP real rates of growth (percent)",np.nan,inplace=True)
```

```python
df.head(50)
```

```python
df1=df.dropna()
```

```python
df1.reset_index(drop=True,inplace=True)
```

```python
df1
```

```python
df.replace("Total, all countries or areas",np.nan,inplace=True)
```

```python
df.head(50)
```

```python
df1=df.dropna()
```

```python
df1.reset_index(drop=True,inplace=True)
```

```python
df1
```

```python
df1.to_csv('C02_GDP_Cleaned.csv')
```
