## Python CSV parser program for C02 Population
#### This program helps to clean the csv files being used for the DV project and only keeps the necessary amout to data and store it in a new CSV file

```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('Population.csv',skiprows=2,names=['Region','CountryName', 'Year', 'Series','Value']  ,usecols=[0,1,2,3,4])
```

```python
df.head(10)
```

```python
df.replace("Population mid-year estimates for males (millions)",np.nan,inplace=True)
df.replace("Population mid-year estimates for females (millions)",np.nan,inplace=True)
df.replace("Sex ratio (males per 100 females)",np.nan,inplace=True)
df.replace("Population aged 0 to 14 years old (percentage)",np.nan,inplace=True)
df.replace("Population aged 60+ years old (percentage)",np.nan,inplace=True)
df.replace("Population density",np.nan,inplace=True)
```

```python
df.head(50)
```

```python
df1=df.dropna()
```

```python
df1.reset_index(drop=True,inplace=True)
```

```python
df.replace("Total, all countries or areas",np.nan,inplace=True)
```

```python
df1.reset_index(drop=True,inplace=True)
```
